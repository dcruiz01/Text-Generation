### Data Manipulation

```python
# read tweet json
df = pd.read_json('/content/drive/My Drive/RESEARCH/TEXT_GENERATION/data/trump_twitter_dataset/trump_original.json')
```

```python
# inspect
print(df.shape)
df.head()
```

```python
# sort by timestamp
df.sort_values('date').head(100)
```

```python
# remove retweets
df_no_retweets = df[df.isRetweet != 't']
print(df_no_retweets.shape)
```

```python
# shuffle rows & reassign indices
df = df.sample(frac = 1).reset_index(drop = True)
```

```python
# split into train & test sets, save to drive
df_test = df[:6694]
df_train = df[6694:]
print(df_test.shape, df_train.shape)
df_test.to_csv('/content/drive/My Drive/RESEARCH/TEXT_GENERATION/data/trump_twitter_dataset/trump_test.csv', index = False)
df_train.to_csv('/content/drive/My Drive/RESEARCH/TEXT_GENERATION/data/trump_twitter_dataset/trump_train.csv', index = False)
```
